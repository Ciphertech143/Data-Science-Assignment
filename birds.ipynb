### Plotting of scatter plot line and bar chat using the above data set

```python
snow_goose_df = df[df['Name'] == 'Snow goose']
# Plotting scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(snow_goose_df['MinBodyMass'], snow_goose_df['MaxLength'], color='blue', label='Data points')
plt.xlabel('Minimum Body Mass')
plt.ylabel('Minimum Wingspan')
plt.title('Scatter Plot of Minimum Body Mass vs Minimum Wingspan for Snow Goose')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting line plot
plt.figure(figsize=(10, 6))
plt.plot(snow_goose_df['MinBodyMass'], snow_goose_df['MaxLength'], color='red', label='Line plot')
plt.xlabel('Minimum Body Mass')
plt.ylabel('Maximum Length')
plt.title('Line Plot of Minimum Body Mass vs Maximum Length for Snow Goose')
plt.legend()
plt.grid(True)
plt.show()
```
